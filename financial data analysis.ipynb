## Price Changes and Financial Returns

__More meaningful/useful than Prices: Price changes__

```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.4f}'.format
```

```python
close = pd.read_csv("close.csv", index_col = "Date", parse_dates = ["Date"])
```

```python
close
```

```python
msft = close.MSFT.dropna().to_frame().copy()
```

```python
msft
```

```python
msft.rename(columns = {"MSFT":"Price"}, inplace = True)
msft
```

```python
msft.shift(periods = 1)
```

```python
msft["P_lag1"] = msft.shift(periods = 1)
msft
```

__Absolute PRice Changes__ (Difference)

__Alternative 1__

```python
msft["P_diff"] = msft.Price.sub(msft.P_lag1)
```

```python
msft
```

__Alternative 2__

```python
msft['P_diff2'] = msft.Price.diff(periods = 1)
msft
```

__-> Absolute Price Changes are not meaninful__

__Relative/Percentage Prie Changes__

```python
msft.Price.div(msft.P_lag1) - 1 # Alternative 1
```

```python
msft["Returns"] = msft.Price.pct_change(periods = 1)
msft
```

__Take Home: Relative Price changes returns are meaningful and comparable across instruments__

```python
msft.drop(columns = ["P_lag1", "P_diff", "P_diff2"], inplace = True)
```

```python
msft
```

```python
msft.to_csv("msft.csv")
```

## Measuring Risk and reward of an investment

__General rule in finance/Investing: Higher Risk must be rewarded with higher returns.__

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.4f}'.format
```

```python
msft = pd.read_csv("msft.csv", index_col = "Date", parse_dates = ["Date"])
msft
```

```python
msft.Price.plot(figsize = (15, 8), fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```

* Reward: Positive Returns
* Risk: Volitility of Returns

```python
msft.describe()
```

```python
mu =  msft.Returns.mean() # arithmetic mean return -> Reward
mu
```

```python
sigma = msft.Returns.std() #standard deviation of returns -> Risk/Volitility
sigma
```
